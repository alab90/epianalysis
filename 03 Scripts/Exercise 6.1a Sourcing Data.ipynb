```python
# checking for mixed types
for col in df_epi_results.columns.tolist():
  weird = (df_epi_results[[col]].applymap(type) != df_epi_results[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_epi_results[weird]) > 0:
    print (col)
```
